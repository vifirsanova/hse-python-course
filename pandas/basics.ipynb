Обработка данных c **аксессорами**


- `df['SentimentText']`: делаем запрос к столбцу `SentimentText`

- `.str`: аксессор, дает доступ к методам класса `str` в питоне

- `.lower()`: метод класса `str`

````python
# приводим к нижнему регистру
df['CleanedText'] = df['SentimentText'].str.lower()
df.head()
````

````python
# примение регулярок
df['CleanedText'] = df['CleanedText'].str.replace(r'[^a-z\s]', '', regex=True)

df.head()
````

Применение функций с **apply**

- `apply(func)`: применяет любую функцию к каждому элементу

Это может быть метод из библиотеки (например, `word_tokenize` из `nltk`) или ваша кастомная функция , которую вы прописали через `def`

````python
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('punkt_tab')

df['Tokens'] = df['CleanedText'].apply(word_tokenize)

df.head()
````

````python
import spacy

nlp = spacy.load('en_core_web_sm')

def extract_entities(text):
    # создаем кастомную функцию, которую также прописываем в apply
    doc = nlp(text)
    return [(ent.text, ent.label_) for ent in doc.ents]

df['Entities'] = df['CleanedText'].apply(extract_entities)

df.head()
````

**lambda-функции**

Так называются одноразовые функции, например

```
lamdbda x: x+1
```

````python
func = lambda x: x+1
func(1)
````

Такие функции занимают мало места в памяти устройства, их легко и быстро писать, поэтому они популярны в big data

````python
from transformers import pipeline

classifier = pipeline('sentiment-analysis')

df['HuggingFace Sentiment'] = df['CleanedText'].apply(lambda x: classifier(x)[0]['label'])

df.head()
````

# фильтрация данных

Объект **NaN**

````python
import numpy as np

np.nan
````

Заменим пустые списки в `Entities` на NaN

````python
df['Entities'] = df['Entities'].apply(lambda x: np.nan if isinstance(x, list) and not x else x)
df.head()
````

Удалим все строчки с пустыми значениями: обычно мы делаем это на этапе data cleaning, т.к. строки с пустыми значениями могут вносить шум в данные, например, в ML

````python
df_cleaned = df.dropna() # df.dropna(subset=['Entities']) - если у нас где-то еще есть NaN, можно сделать фильтрацию по отдельным столбцам
df_cleaned
````

````python
# метод 2: фильтрация по индексу, по столбцами
df_cleaned.iloc[:, :3]
````

````python
# метод 3: фильтрация по индексу, по ячейкам
df_cleaned.iloc[[0, 2], [1, 3]]
````
