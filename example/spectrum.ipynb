```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from matplotlib import mlab
```

```python
sample_rate, audio_data = wavfile.read('ImperialMarch60.wav')
```

```python
sample_rate # часотота дискретизации в Гц
```

```python
audio_data # амплитуда
```

```python
plt.figure(figsize=(10, 6))

plt.specgram(audio_data, Fs=sample_rate, NFFT=1024, noverlap=512, cmap='viridis')

plt.xlabel('Time [s]')
plt.ylabel('Frequency [Hz]')
plt.title('Spectrogram of the Audio Signal')
plt.colorbar(label='Intensity [dB]')

plt.show()
```

`Fs=sample_rate`: Частота дискретизации

`NFFT=1024`: количество блоков данных для формирования окна для преобразований Фурье

`noverlap=512`: пересечение между блоками, повышает непрерывность
