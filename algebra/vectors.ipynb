Ищем расстояние между векторами

1. Задаем пространство

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

plt.show()
```

2. Задаем 2 вектора

```python
vector_a = np.array([0.01, 0.02, 0.04])
vector_b = np.array([-0.04, -0.04, -0.04])

vector_a, vector_b
```

3. Задаем векторное пространство

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(0, 0, 0, vector_a[0], vector_a[1], vector_a[2], color='r', label='Vector A', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, vector_b[0], vector_b[1], vector_b[2], color='b', label='Vector B', arrow_length_ratio=0.1)

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Vector Visualization')
ax.legend()

plt.show()
```

Манэттенское расстояние

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(0, 0, 0, vector_a[0], vector_a[1], vector_a[2], color='r', label='Vector A', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, vector_b[0], vector_b[1], vector_b[2], color='b', label='Vector B', arrow_length_ratio=0.1)

l1_line = np.array([vector_a, [vector_a[0], vector_b[1], vector_a[2]], vector_b])
ax.plot(l1_line[:, 0], l1_line[:, 1], l1_line[:, 2], color='g', linestyle='--', label='L1 Distance')

l1_distance = np.sum(np.abs(vector_a - vector_b))
plt.figtext(0.15, 0.85, f'L1 Distance: {l1_distance}', fontsize=10)

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Vector Visualization with L1 Distance')
ax.legend()

plt.show()
```

Евклидово расстояние

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(0, 0, 0, vector_a[0], vector_a[1], vector_a[2], color='r', label='Vector A', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, vector_b[0], vector_b[1], vector_b[2], color='b', label='Vector B', arrow_length_ratio=0.1)

l1_line = np.array([vector_a, [vector_a[0], vector_b[1], vector_a[2]], vector_b])
ax.plot([vector_a[0], vector_b[0]], [vector_a[1], vector_b[1]], [vector_a[2], vector_b[2]], color='orange', linestyle=':', label='L2 Distance')

l2_distance = np.linalg.norm(vector_a - vector_b)
plt.figtext(0.15, 0.85, f'L2 Distance: {l2_distance}', fontsize=10)

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Vector Visualization with L2 Distance')
ax.legend()

plt.show()
```

Скалярное произведение

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(0, 0, 0, vector_a[0], vector_a[1], vector_a[2], color='r', label='Vector A', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, vector_b[0], vector_b[1], vector_b[2], color='b', label='Vector B', arrow_length_ratio=0.1)

dot_product = np.dot(vector_a, vector_b)
plt.figtext(0.15, 0.85, f'Dot Product: {dot_product}', fontsize=10)

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Vector Visualization with Dot Product')
ax.legend()

plt.show()
```

Косинусное сходство

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(0, 0, 0, vector_a[0], vector_a[1], vector_a[2], color='r', label='Vector A', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, vector_b[0], vector_b[1], vector_b[2], color='b', label='Vector B', arrow_length_ratio=0.1)

cosine_similarity = dot_product / (np.linalg.norm(vector_a) * np.linalg.norm(vector_b))

plt.figtext(0.15, 0.85, f'Cosine Similarity: {cosine_similarity}', fontsize=10)

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Vector Visualization with Cosine Similarity')
ax.legend()

plt.show()
```

Регуляризация

```python
vector_a = np.array([0.01, 0.02, 0.04])
vector_b = np.array([-1, -6, -10])

vector_a, vector_b
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(0, 0, 0, vector_a[0], vector_a[1], vector_a[2], color='r', label='Vector A', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, vector_b[0], vector_b[1], vector_b[2], color='b', label='Vector B', arrow_length_ratio=0.1)

l1_line = np.array([vector_a, [vector_a[0], vector_b[1], vector_a[2]], vector_b])
ax.plot(l1_line[:, 0], l1_line[:, 1], l1_line[:, 2], color='g', linestyle='--', label='L1 Distance')
ax.plot([vector_a[0], vector_b[0]], [vector_a[1], vector_b[1]], [vector_a[2], vector_b[2]], color='orange', linestyle=':', label='L2 Distance')

l1_distance = np.sum(np.abs(vector_a - vector_b))
plt.figtext(0.50, 0.85, f'L1 Distance: {l1_distance}', fontsize=10)
plt.figtext(0.80, 0.85, f'L2 Distance: {l2_distance}', fontsize=10)

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Vector Visualization with L1 and L2 Distances')
ax.legend()

plt.show()
```

```python
vector_a_l1 = tf.linalg.normalize(vector_a, ord=1)
vector_b_l1 = tf.linalg.normalize(vector_b, ord=1)

vector_a_l1, vector_b_l1
```

```python
vector_a_l2 = tf.linalg.normalize(vector_a, ord=2)
vector_b_l2 = tf.linalg.normalize(vector_b.astype('float64'), ord=2)
vector_a_l2, vector_b_l2
```

```python
vector_a, vector_b = vector_a_l1[0], vector_b_l1[0]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(0, 0, 0, vector_a[0], vector_a[1], vector_a[2], color='r', label='Vector A', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, vector_b[0], vector_b[1], vector_b[2], color='b', label='Vector B', arrow_length_ratio=0.1)

l1_line = np.array([vector_a, [vector_a[0], vector_b[1], vector_a[2]], vector_b])
ax.plot(l1_line[:, 0], l1_line[:, 1], l1_line[:, 2], color='g', linestyle='--', label='L1 Distance')
ax.plot([vector_a[0], vector_b[0]], [vector_a[1], vector_b[1]], [vector_a[2], vector_b[2]], color='orange', linestyle=':', label='L2 Distance')

l1_distance = np.sum(np.abs(vector_a - vector_b))
plt.figtext(0.50, 0.85, f'L1 Distance: {round(l1_distance, 5)}', fontsize=10)
plt.figtext(0.80, 0.85, f'L2 Distance: {round(l2_distance, 5)}', fontsize=10)

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Vector Visualization with L1 and L2 Distances')
ax.legend()

plt.show()
```

```python
vector_a, vector_b = vector_a_l2[0], vector_b_l2[0]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(0, 0, 0, vector_a[0], vector_a[1], vector_a[2], color='r', label='Vector A', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, vector_b[0], vector_b[1], vector_b[2], color='b', label='Vector B', arrow_length_ratio=0.1)

l1_line = np.array([vector_a, [vector_a[0], vector_b[1], vector_a[2]], vector_b])
ax.plot(l1_line[:, 0], l1_line[:, 1], l1_line[:, 2], color='g', linestyle='--', label='L1 Distance')
ax.plot([vector_a[0], vector_b[0]], [vector_a[1], vector_b[1]], [vector_a[2], vector_b[2]], color='orange', linestyle=':', label='L2 Distance')

l1_distance = np.sum(np.abs(vector_a - vector_b))
plt.figtext(0.50, 0.85, f'L1 Distance: {round(l1_distance, 5)}', fontsize=10)
plt.figtext(0.80, 0.85, f'L2 Distance: {round(l2_distance, 5)}', fontsize=10)

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Vector Visualization with L1 and L2 Distances')
ax.legend()

plt.show()
```
